```python
# creating empty lists for scraping data
Repository_title=[]
Repository_description=[]
```

```python
Contributors_count=[]
Language_used=[]
```

```python
#accessing full details of repositories
urls=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a") #collecting urls of all the repositories
UR=[]
for i in urls:
    UR.append(i.get_attribute('href'))#getting the urls into list
for url in UR:#loop for every repository in the list    

    driver.get(url)
    try: 
        title=driver.find_element_by_xpath("//strong[@class='mr-2 flex-self-stretch']/a")#locating the title
        Repository_title.append(title.text)#appending the title in list
    except NoSuchElementException as e:
        Repository_title.append("--")
    try:     
        description=driver.find_element_by_xpath("//p[@class='f4 mt-3']")# locating Repository_description
        Repository_description.append(description.text)#appending the description in list
    except NoSuchElementException as e:
        Repository_description.append("--")
```

```python
for url in UR:#loop for every repository in the list
    
    driver.get(url)
    try:
        element=driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div[1]/h2[1]/a/span")
        #contributors=driver.find_elements_by_class_name("BorderGrid BorderGrid--spacious")#locating the Contributors
        if(element.text is not None):
            Contributors_count.append(element.text)
        else:
            element2=driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div[1]/h2[1]/a/span")
            Contributors_count.append(element2.text)    
    except NoSuchElementException as e:
        Contributors_count.append("--")
```

```python
Contributors_count
```

```python
for url in UR:#loop for every repository in the list    

    driver.get(url)
    try:
        language=driver.find_elements_by_xpath("//div[@class='BorderGrid-cell']/ul/li/a/span[1]")#locating the languages
        for i in language:
            Language_used.append(i.text)#appending the language in list
    except NoSuchElementException as e:
        Language_used.append("--")
```

```python
Language_used
```

```python
GitHub_Repositories={}
GitHub_Repositories['Repository_title'] = Repository_title
GitHub_Repositories['Repository_description'] = Repository_description
GitHub_Repositories['Contributors_count'] = Contributors_count
GitHub_Repositories['Language_used'] = Language_used
```
